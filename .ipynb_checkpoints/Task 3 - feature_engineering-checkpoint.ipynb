## 3. Feature engineering

### Difference between off-peak prices in December and preceding January

Below is the code created by your colleague to calculate the feature described above. Use this code to re-create this feature and then think about ways to build on this feature to create features with a higher predictive power.

```python
price_df = pd.read_csv('price_data.csv')
price_df["price_date"] = pd.to_datetime(price_df["price_date"], format='%Y-%m-%d')
price_df.head()
```

```python
# Group off-peak prices by companies and month
monthly_price_by_id = price_df.groupby(['id', 'price_date']).agg({'price_off_peak_var': 'mean', 'price_off_peak_fix': 'mean'}).reset_index()

# Get january and december prices
jan_prices = monthly_price_by_id.groupby('id').first().reset_index()
dec_prices = monthly_price_by_id.groupby('id').last().reset_index()

# Calculate the difference
diff = pd.merge(dec_prices.rename(columns={'price_off_peak_var': 'dec_1', 'price_off_peak_fix': 'dec_2'}), jan_prices.drop(columns='price_date'), on='id')
diff['offpeak_diff_dec_january_energy'] = diff['dec_1'] - diff['price_off_peak_var']
diff['offpeak_diff_dec_january_power'] = diff['dec_2'] - diff['price_off_peak_fix']
diff = diff[['id', 'offpeak_diff_dec_january_energy','offpeak_diff_dec_january_power']]
diff.head()
```

Now it is time to get creative and to conduct some of your own feature engineering! Have fun with it, explore different ideas and try to create as many as yo can!

```python
#let merge the dataset
data = price_df.merge(diff, on='id')
df = data.merge(df, on='id')
df.head()
```

```python
#let's check the general info of our new dataframe
df.info()
```

```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(32,26))
mask = np.triu(np.ones_like(df.corr(), dtype=bool))
sns.heatmap(df.corr(), mask = mask, center=0, cmap ="RdBu", linewidths=1, annot=True, fmt=".2f", vmin=-1, vmax=1)
plt.show()
```

```python
#let's drop columns that have strong correlation
df= df.drop(['var_6m_price_off_peak_var','var_6m_price_peak_var','var_6m_price_mid_peak_var','var_6m_price_off_peak_fix','var_6m_price_peak_fix','var_6m_price_mid_peak_fix','var_6m_price_off_peak','var_6m_price_peak','var_6m_price_mid_peak','var_year_price_mid_peak_var','var_year_price_mid_peak_fix','var_year_price_mid_peak','price_mid_peak_var','price_mid_peak_fix','margin_gross_pow_ele','forecast_cons_year','forecast_price_pow_off_peak','forecast_cons_12m','forecast_price_energy_peak','forecast_price_energy_off_peak','cons_last_month','forecast_meter_rent_12m','var_year_price_off_peak','var_year_price_peak'], axis=1)
```

```python
#let's convert data types to the right datatype
df['has_gas'] = df['has_gas'].replace({'t':True,'f':False}).astype(bool)
df['origin_up'] = df['origin_up'].astype('category')
df['channel_sales'] = df['channel_sales'].astype('category')
```

```python
#let's create a variable for the overall consumption
df['total_consumption_12m'] = df['cons_12m'] + df['cons_gas_12m']
#let's calculate the tenure
df['tenure'] = ((df['date_end'] - df['date_activ'])/np.timedelta64(1,'Y')).astype(int)
```

```python
#let's create day month and year based on date
df['activ_month'] = df['date_activ'].dt.month
df['activ_day'] = df['date_activ'].dt.day
df['activ_year'] = df['date_activ'].dt.year
df['end_month'] = df['date_end'].dt.month
df['end_day'] = df['date_end'].dt.day
df['end_year'] = df['date_end'].dt.year
df['modif_prod_month'] = df['date_modif_prod'].dt.month
df['modif_prod_day'] = df['date_modif_prod'].dt.day
df['modif_prod_year'] = df['date_modif_prod'].dt.year
df['renewal_month'] = df['date_renewal'].dt.month
df['renewal_day'] = df['date_renewal'].dt.day
df['renewal_year'] = df['date_renewal'].dt.year
```

```python
#let's now remove date columns
df = df.drop(['date_activ','date_end','date_modif_prod','date_renewal'], axis=1)
```

```python
#the distribution of the dataset
for n in df.select_dtypes(include=['int','float']):
    plt.figure()
    sns.displot(df[n], kind='kde')
    plt.title(f"Distribution of {n}")
```

```python
#let's apply log transformation to skewed columns
for c in ['offpeak_diff_dec_january_power','cons_12m','cons_gas_12m','forecast_discount_energy','imp_cons','margin_net_pow_ele','nb_prod_act','net_margin','pow_max','var_year_price_off_peak_var','var_year_price_peak_var','var_year_price_off_peak_fix','var_year_price_peak_fix','total_consumption_12m']:
    df[c] = np.log10(df[c] +1)
```

```python
#let's one hot encode the categorical variable
df = pd.get_dummies(df, columns=['channel_sales','origin_up'])
```
